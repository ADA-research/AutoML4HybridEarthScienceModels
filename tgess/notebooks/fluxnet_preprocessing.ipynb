# FLUXNET Pre-processing

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
data_path = "../data/raw/gpp"
site_folder = "FLX_DE-RuS_FLUXNET2015_SUBSET_2011-2014_1-4"
daily_file = "FLX_DE-RuS_FLUXNET2015_SUBSET_DD_2011-2014_1-4.csv"
```

```python
df = pd.read_csv(os.path.join(data_path, site_folder, daily_file))
```

```python
def timestamp_to_date(x):
    return datetime.strptime(str(x), "%Y%M%d")
```

```python
df["date"] = df["TIMESTAMP"].apply(lambda x: datetime.strptime(str(x), "%Y%M%d").date())
```
